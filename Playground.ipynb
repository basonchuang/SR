```python
import numpy as np
import scipy.io.wavfile as wav
from python_speech_features import mfcc
import librosa, os
from keras.utils import to_categorical
```

```python
path = "./dataset/data_sr/x/"

def get_labels(path):
    labels = os.listdir(path)
    label_indices = np.arange(0, len(labels))
    return labels, label_indices, to_categorical(label_indices)
```

```python
max_pad_len = 68
```

```python
def get_max_pad_len(path):
    labels, _, _ = get_labels(path)
    max_pad_len = 0
    
    for label in labels:
        wavfiles = [path + label + '/' + wavfile for wavfile in os.listdir(path + '/' + label)]
        for wavfile in wavfiles:
            wave, sr = librosa.load(wavfile, mono=True, sr=None)
            mfcc = librosa.feature.mfcc(y=wave, sr=sr)
            if mfcc.shape[1] > max_pad_len:
                max_pad_len = mfcc.shape[1]
                
        print(label + "Done")
        
    return max_pad_len

max_pad_len = get_max_pad_len(path)
                
print(max_pad_len)
```

```python


def wav2mfcc(file_path, max_pad_len):
    try:
        wave, sr = librosa.load(file_path, mono=True, sr=None)
        mfcc = librosa.feature.mfcc(y=wave, sr=sr)
        
        pad_width = max_pad_len - mfcc.shape[1]
        if pad_width < 0:
            pad_width = 0
            mfcc = mfcc[:,:max_pad_len]
        mfcc = np.pad(mfcc, pad_width=((0, 0), (0, pad_width)), mode='constant')
        return mfcc
        
    except:
        print(file_path)


labels, _, _ = get_labels(path)
```

```python
for label in labels:
    mfcc_vectors = []
    
    wavfiles = [path + label + '/' + wavfile for wavfile in os.listdir(path + '/' + label)]
    
    count = 5
    
    for wavfile in wavfiles:
        mfcc = wav2mfcc(wavfile, max_pad_len=128)
        mfcc_vectors.append(mfcc)
        count -=1
        
print(mfcc_vectors[0].shape)
```
